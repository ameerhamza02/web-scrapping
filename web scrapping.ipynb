```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as cl
```

```python
url='https://www.flipkart.com/search?q=earbuds&sid=0pm%2Cfcn&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=earbuds%7CHeadphones&requestId=bec99587-86b1-4888-a9a8-d4e1b0a39a88&as-searchtext=ear'
```

```python
res=requests.get(url)
```

```python
res
```

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
earbuds_name=soup.select('.s1Q9rs')
```

```python
earbuds_name
```

```python
earbuds_price=soup.select('._8VNy32 ._30jeq3')
```

```python
earbuds_price
```

```python
earbuds_offer=soup.select('._8VNy32 span')
```

```python
earbuds_offer
```

```python
df=cl.DataFrame(earbuds_name,columns=["earbuds_name"])
```

```python
df
```

```python
df['earbuds_price']=earbuds_price
```

```python
df
```

```python
df['earbuds_offer']=earbuds_offer
```

```python
df
```

```python
df.to_csv('amir.csv')
```
